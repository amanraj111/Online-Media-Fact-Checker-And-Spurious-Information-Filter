# **Data Preprocessing**

Taking care of missing data

```
# checking missing data in each column of training set
dataset.isnull().sum()
```

```
# Creating a new column with name total
dataset['total'] = dataset['title'] + ' ' + dataset['text']
print(dataset.shape)
dataset
```

Taking care of categorical variables

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

Cleaning Text and stemming

```
# for training set
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(dataset.shape[0]):
  news = re.sub('[^a-zA-Z]', ' ', dataset.iloc[i,-1])
  news = news.lower()
  news = news.split()
  all_stopwords = stopwords.words('english')
  ps = PorterStemmer()
  lis = []
  for word in news:
    if word not in all_stopwords:
      word = ps.stem(word)
      lis.append(word)
  corpus.append(" ".join(lis))
```

```
corpus[0]
```

# **Tf-IDF Vectorizer**

```
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features= 20000,ngram_range=(1,3))
x = tfidf.fit_transform(corpus).toarray()
x.shape
```

```
count_df = pd.DataFrame(x, columns=tfidf.get_feature_names())
count_df.head(50)
```

# Splitting the training and test set

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=0)
x_train.shape
```

MultinomialNB

```
#DataFlair - Initialize a PassiveAggressiveClassifier
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(x_train, y_train)
```

Passive Aggresive Classifier

```
from sklearn.linear_model import PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(x_train, y_train)
```

Predicting the result

```
# predicting result for multinomialNB
y_pred_m = classifier.predict(x_test)
y_pred_m
```

```
# predicting result for passiveagressive
y_pred_p = pac.predict(x_test)
y_pred_p
```

Confusion Matrix and Accuracy Score

```
# for multinomialnb
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred_m)
print(cm)
accuracy_score(y_test, y_pred_m)
```

```
# for passive agressive Classifier
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred_p)
print(cm)
accuracy_score(y_test, y_pred_p)
```
